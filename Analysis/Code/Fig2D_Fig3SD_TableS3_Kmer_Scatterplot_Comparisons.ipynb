## Figure 2D; Supplementary Figure 3D; Supplementary Table 3B-D

Analysis for scatterplots comparing k-mers among non_UV and UV conditions.


### Files

K-mer files indicated in the meta data file: Meta_Data_Kmer_Comparisons.csv. 


### Overview - K-mer comparisons across conditions

**Supplementary Table 3B: Z-scores**

For each of the 10 TFs, the 7-mer **Z-scores** for non-UV and UV conditions were placed in a wide-format table. These are saved in a csv format table.

**Supplementary Table 3C: E-scores**

For each of the 10 TFs, the 7-mer **E-scores** for non-UV and UV conditions were placed in a wide-format table. These are saved in a csv format table.

**Supplementary Table 3D: Z-scores Slope**

For each comparison, the slope of the k-mers that are above the threshold in both conditions was calculated along with a 95% confidence interval. These are saved in a csv format table.

**Supplementary Table 3E: K-mer Group Percentages**:

Data shown in Supplementary Figure 3D. This is the percentage of each group (Both conditions below threshold, one condition above threshold, both conditions above threshold) that is used to color the data points in Figure 2D. 

**Figure 2D: Scatterplots of Z-scores**

For each of the 10 transcription factors, the 7-mer E-scores and Z-scores comparing binding to non-irradiated and UV-irradiated DNA are plotted as scatterplots. Plots are colored by an E-score threshold of 0.35 to indicate if 7-mers are above the threshold in both, one, or none of the conditions. The output is a png format plot containing all scatterplots as subpanels. 

**Supplementary Figure 3D: Barplot of K-mer Group Percentages**

For the 10 transcription factors, a stacked barplot of the percentages of each k-mer group (Both below threshold, one above threshold, both above threshold) is created along with a zoomed-in version. 

**Note:**

Files output in **csv** format for this notebook were manually placed and organized in the **xlsx** format supplementary table.

```python
# Imports
#import math
import os

from bokeh.io import export_svg
from bokeh.models import ColumnDataSource
from bokeh.plotting import figure, show
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from scipy import stats
from scipy.stats import t


import uvbind_analysis_core as uac

# Global variables
META_DATA_KMERS = "Meta_Data/Meta_Data_Kmer_Comparisons.csv"
DATA_FOLDER_KMERS = "../../Data/Kmer_Files"
# Plot settings 
THRESHOLD = 0.35
COLOR_PALETTE = ("black" ,"#636363","#bdbdbd")
FONT = "arial"
FIGURE_2_OUTPUT = "../Figure_2"
TABLE_S3_OUTPUT = "../Table_S3"
OUTPUT_FOLDER_S3 = "../Figure_S3"
GROUPS = ["Both_Above", "One_Each", "Both_Below"]
CIRCLE_SIZE = 70
TOTAL_6MERS = 8192
```

### Supplementary Table B-C



#### (1) Read and display meta data:

```python
kmer_meta = pd.read_csv(META_DATA_KMERS)
os.makedirs(FIGURE_2_OUTPUT, exist_ok=True)
os.makedirs(TABLE_S3_OUTPUT, exist_ok=True)
kmer_meta
```

#### (2) Analysis - Create and save tables of E-scores and Z-scores:

```python
dfs = []
for row in kmer_meta.itertuples():
    df = pd.read_csv(f"{DATA_FOLDER_KMERS}/{row.Non_UV_File}", sep='\t')
    df["Experiment"] = row.Non_UV_File.split("_")[0] + "_NonUV"
    dfs.append(df)
    df = pd.read_csv(f"{DATA_FOLDER_KMERS}/{row.UV_File}", sep='\t')
    df["Experiment"] = "_".join(row.UV_File.split("_")[:2])
    dfs.append(df)
result = pd.concat(dfs)
# S table of z-scores
zscore_df = result[["7-mer", "7-mer.1", "Z-score", "Experiment"]]
zscore_df = zscore_df.pivot(index=["7-mer", "7-mer.1"], columns="Experiment", values="Z-score")
zscore_df = zscore_df.reset_index()
zscore_df.to_csv(f"{TABLE_S3_OUTPUT}/Table_S3B_Zscores.csv", index=False)
# S table of E-scores
escore_df = result[["7-mer", "7-mer.1", "E-score", "Experiment"]]
escore_df = escore_df.pivot(index=["7-mer", "7-mer.1"], columns="Experiment", values="E-score")
escore_df = escore_df.reset_index()
escore_df.to_csv(f"{TABLE_S3_OUTPUT}/Table_S3C_Escores.csv", index=False)
```

### Figure 2D, Supplementary Table 3D



#### (1) Functions

```python
def group_on_threshold(dataframe: pd.DataFrame,
                       column_a,
                       column_b,
                       threshold: float = THRESHOLD,
                       groups: list = GROUPS):
    """Add a color column based on a threshold value.

    Given a k-mer dataframe, names of column a and b, a threshold value,
    and a 3 color palette, goes over every row of the dataframe and
    selects a color based on if both, one, or none of columns a and b
    are above the threshold. Returns a dataframe with a Color column of
    the selected colors.
    """
    result = []
    for a, b in zip(dataframe[column_a], dataframe[column_b]):
        if a >= threshold and b >= threshold:
            result.append(groups[0])
        elif a >= threshold or b >= threshold:
            result.append(groups[1])
        else:
            result.append(groups[2])
    return result

def zscore_scatterplot(x,
                       y,
                       group,
                       ax,
                       plot_start,
                       plot_end,
                       plot_df,
                       circle_size):
    """Add a scatterplot to an ax object in a matplotlib figure."""
    sns.scatterplot(x=x,
                    y=y,
                    data=plot_df,
                    ax=ax,
                    linewidth=0,
                    hue = group,
                    hue_order = GROUPS,
                    palette = COLOR_PALETTE,
                    legend=False,
                    s=circle_size)
    ax.set_xlim(plot_start, plot_end)
    ax.set_ylim(plot_start, plot_end)
    ax_array[idx].set_xticks([])
    ax_array[idx].set_yticks([])
    ax_array[idx].set_xticklabels([])
    ax_array[idx].set_yticklabels([])
    ax_array[idx].set_ylabel("")
    ax_array[idx].set_xlabel("")
    ax_array[idx].axline((0, 0), slope=1, color="black")
    ax_array[idx].set(adjustable='box', aspect='equal')
```

#### (2) Analysis - Supplementary Table S3D

```python
# Group based on E-score
kmer_groups = escore_df[["7-mer", "7-mer.1"]].copy()
for protein in kmer_meta["Protein"]:
    kmer_groups[f"{protein}_Group"] = group_on_threshold(escore_df, f"{protein}_NonUV", f"{protein}_UV")
plot_df = pd.merge(zscore_df, kmer_groups, on=["7-mer", "7-mer.1"])
tinv = lambda p, df: abs(t.ppf(p/2, df))
# Linear regression for ordering scatterplots
slopes = []
confidence_intervals = []
for row in kmer_meta.itertuples():
    sig_kmers = plot_df.copy()
    #sig_kmers = sig_kmers[sig_kmers[f"{row.Protein}_Group"] != "Both_Below"].reset_index(drop=True)
    sig_kmers = sig_kmers[sig_kmers[f"{row.Protein}_Group"] == "Both_Above"].reset_index(drop=True)
    ts = tinv(0.05, len(sig_kmers)-2)
    reg = stats.linregress(sig_kmers[f"{row.Protein}_NonUV"], sig_kmers[f"{row.Protein}_UV"])
    slopes.append(reg.slope)
    confidence_intervals.append(ts*reg.stderr)
kmer_meta["Slope_Sig"] = slopes
kmer_meta["CI_95_Sig"] = confidence_intervals
kmer_meta = kmer_meta.sort_values(by="Slope_Sig")
kmer_meta.to_csv(f"{TABLE_S3_OUTPUT}/Table_S3D_Zscore_Slopes.csv", index=False)
```

#### (3) Analysis - Plot Figure 2D

```python
# Plot
plot_count = len(kmer_meta)
fig, ax_array = plt.subplots(1, plot_count)
fig.set_size_inches(plot_count ** 2, plot_count)
for idx, row in enumerate(kmer_meta.itertuples()):
    zscore_scatterplot(f"{row.Protein}_NonUV",
                       f"{row.Protein}_UV",
                       f"{row.Protein}_Group",
                       ax_array[idx],
                       row.Plot_Start,
                       row.Plot_End,
                       plot_df,
                       CIRCLE_SIZE)
fig.savefig(fname=f"{FIGURE_2_OUTPUT}/Fig2D_Zscore_Plots.png", format='png')
```

### Supplementary Figure 3D, Supplementary Table 3

#### (1) Analysis

```python
dfs = []
for row in kmer_meta.itertuples():
    df = plot_df[f"{row.Protein}_Group"].value_counts().reset_index()
    df = df.rename(columns={"index":"Category",f"{row.Protein}_Group":"Counts"})
    df["Protein"] = f"{row.Protein}"
    dfs.append(df)
result = pd.concat(dfs).reset_index()

barplotdf = result.pivot(index="Protein", values="Counts", columns="Category").reset_index()
barplotdf = barplotdf.sort_values(by="Both_Below").reset_index()
for i in ["Both_Below","One_Each", "Both_Above" ]:
    barplotdf[i] = (barplotdf[i] / TOTAL_6MERS) * 100
barplotdf = barplotdf[["Protein", "Both_Above", "One_Each", "Both_Below"]]
barplotdf.to_csv(f"{TABLE_S3_OUTPUT}/Group_Percentages.csv", index=False,  float_format='%.2f')
```

#### (2) Plot

```python
# Full stacked barplot
p = figure(x_range=barplotdf["Protein"], height=250 , y_range=(0, 105))

p.vbar_stack(["Both_Below","One_Each", "Both_Above" ],
             x="Protein",
             width=0.9, color=COLOR_PALETTE[::-1], source=barplotdf)

p.x_range.range_padding = 0.1
p.xgrid.grid_line_color = None
p.axis.minor_tick_line_color = None
p.outline_line_color = None
show(p)
export_svg(p, filename=f"{OUTPUT_FOLDER_S3}/FigS3D_Kmer_Barplot.svg")
```
