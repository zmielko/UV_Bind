#### (1) Functions

```python
def bin_by_jnb(binby_1d_data, percentile=100, bins=3):
    """Returns bins using Jenks Natural Breaks.
    
    Give a 1d numeric input dataset a percentile, and number of bins,
    subsets the input dataset by the percentile for min and max values
    and returns bins with +/- infinite outer bins.
    """
    # Calculate the min and max value to use for binning
    min_percentile = 100 - percentile
    min_value = np.percentile(binby_1d_data, min_percentile)
    max_value = np.percentile(binby_1d_data, percentile)
    # Adjust the input data used based on the percentile
    binby_adj = list(filter(lambda x: min_value <= x <= max_value, binby_1d_data))
    # Bin using Jenks Natural Breaks
    jnb = JenksNaturalBreaks(bins)
    jnb.fit(binby_adj)
    # Use +/- infinite as the outer breaks
    result = [-math.inf] + jnb.inner_breaks_ + [math.inf]
    return result

def bin_by_range(binby_1d_data, percentile, nbin=3):
    """Returns bins using a range of values."""
    min_percentile = 100 - percentile
    min_value = np.percentile(binby_1d_data, min_percentile)
    max_value = np.percentile(binby_1d_data, percentile)
    bin_size = (max_value - min_value) / nbin
    result = [-math.inf]
    new_bin = min_value
    for i in range(nbin - 1):
        new_bin = new_bin + bin_size
        result.append(new_bin)
    result.append(math.inf)
    return tuple(result)

def ranksum_comparison(df, value_column, groups):
    """Perform a Wilcoxon rank sum test with FDR correction."""
    test_results= []
    # For each group, find the wilxocon rank-sum results
    for group_a, group_b in itertools.combinations(groups, 2):
        values_a = df[df["Effect_Group"] == group_a][value_column]
        values_b = df[df["Effect_Group"] == group_b][value_column]
        p_value = stats.ranksums(values_a, values_b).pvalue
        test_results.append((group_a, group_b, p_value))
    result_df = pd.DataFrame(test_results)
    result_df = result_df.rename(columns={0:"Group_A", 1:"Group_B", 2:"P_Value"})
    fdr_values = multipletests(result_df["P_Value"], method="fdr_bh")[1]
    result_df["FDR"] = fdr_values
    return result_df


def pipeline_competition_figure(competition_df,
                                bin_technique,
                                tf_nuv_col,
                                tf_uv_col,
                                x_ticks_scatter,
                                y_ticks_scatter,
                                y_ticks_uv_bxplt,
                                y_ticks_nuv_bxplt,
                                xlim_scatter,
                                ylim_scatter,
                                ylim_uv_bxplt,
                                ylim_nuv_bxplt,
                                plot_output,
                                statistics_output,
                                nbin=3,
                                palette=PALETTE,
                                bin_percentile=100,
                                show_stats=True,
                                repair_col="UVDDB",
                                repair_competition_col="UVDDB_ETS1"):
    competition_df["Competition_Effect"] = competition_df["UVDDB"] - competition_df["UVDDB_ETS1"]
    if bin_technique == "by_jnb":
        bins = bin_by_jnb(competition_df["Competition_Effect"], bin_percentile, nbin)
    elif bin_technique == "by_range":
        bins = bin_by_range(competition_df["Competition_Effect"], bin_percentile, nbin)
    else:
        raise ValueError("bin_technique must be 'by_jnb' or 'by_range'")
    competition_df["Effect_Group"] = pd.cut(competition_df["Competition_Effect"],
                                            bins,
                                            labels=list(palette.keys()))
    competition_df["Color"] = competition_df["Effect_Group"].apply(lambda x: palette[x])
    # Figure parameters and setup
    plt.rcParams["figure.figsize"] = [20,8] # Wide aspect ratio
    # 3 subplots, with the scatterplot being larger by 3:1
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, gridspec_kw={'width_ratios': [3, 2, 2]})
    plt.subplots_adjust(wspace=0.3) # Adjust space between plots 
    # Draw scatterplot
    ax1.scatter(competition_df[repair_col],
                competition_df[repair_competition_col],
                c=competition_df["Color"])
    ax1.axline((1, 1), slope=1, color='black') # Draw the diagonal
    # Draw the boxplots for non-UV and UV signal
    for bxplt_ax, y_axis in ((ax2, tf_uv_col), (ax3, tf_nuv_col)):
        sns.boxplot(x="Effect_Group",
                    y=y_axis,
                    showfliers=False,
                    order=palette.keys(),
                    palette=palette.values(),
                    data=competition_df,
                    saturation=1,
                    ax=bxplt_ax)
        if show_stats:
            stats = ranksum_comparison(competition_df, y_axis, list(palette.keys()))
            stats["FDR"] = stats["FDR"].apply(lambda x: "FDR = " + '{:.2e}'.format(x))
            #stats["FDR"] = stats["FDR"].apply(lambda x: "*" if x <=0.05 else "n.s.")
            stat_dict = dict(zip(zip(stats["Group_A"], stats["Group_B"]), stats["FDR"]))
            #bxplt_ax.set_ylim()
            # Draw annotations
            annot = Annotator(bxplt_ax, [("Weak", "Medium"), ("Weak", "Strong"), ("Medium", "Strong")],
                              data=competition_df, x="Effect_Group", y=y_axis, order=palette.keys())
            annot.set_custom_annotations([stat_dict[("Weak", "Medium")],
                                          stat_dict[("Weak", "Strong")],
                                          stat_dict[("Medium", "Strong")],])
            bxplt_ax, test_results = annot.annotate()
        #print(test_results)
    # Adjust tick marks and plot draw limits
    ax1.set_xticks(x_ticks_scatter)
    ax1.set_yticks(y_ticks_scatter)
    ax1.set_yticklabels(y_ticks_scatter, rotation=90, va="center")
    ax2.set_yticks(y_ticks_uv_bxplt)
    ax2.set_yticklabels(y_ticks_uv_bxplt, rotation=90, va="center")
    ax3.set_yticks(y_ticks_nuv_bxplt)
    ax3.set_yticklabels(y_ticks_nuv_bxplt, rotation=90, va="center")
    ax1.set_xlabel('UV-DDB Alone\n (UV-DDB Signal; a.u.)', fontsize=20)
    ax1.set_ylabel('UV-DDB + ETS1\n (UV-DDB Signal; a.u.)', fontsize=20)
    ax2.set_ylabel('ETS1 UV Signal (a.u.)', fontsize=20)
    ax2.set_xlabel('', fontsize=0)
    ax3.set_xlabel('', fontsize=0)
    ax3.set_ylabel('ETS1 Non-UV Signal (a.u.)', fontsize=20)
    ax1.set_xlim(xlim_scatter[0], xlim_scatter[1])
    ax1.set_ylim(ylim_scatter[0], ylim_scatter[1])
    matplotlib.rc('xtick', labelsize=20) 
    matplotlib.rc('ytick', labelsize=20)
    #ax2.set_ylim(ylim_uv_bxplt[0], ylim_uv_bxplt[1])
    #ax3.set_ylim(ylim_nuv_bxplt[0], ylim_nuv_bxplt[1])
    plt.savefig(fname=plot_output, format='svg')
    # Run statistics
    stat_groups = list(palette.keys())
    non_uv_statistics_df = ranksum_comparison(competition_df, tf_nuv_col, stat_groups)
    non_uv_statistics_df["Column"] = tf_nuv_col
    uv_statistics_df = ranksum_comparison(competition_df, tf_uv_col, stat_groups)
    uv_statistics_df["Column"] = tf_uv_col
    statistics_df = pd.concat([non_uv_statistics_df, uv_statistics_df]).reset_index(drop=True)
    statistics_df.to_csv(f"{OUTPUT_FOLDER_MAIN}/{statistics_output}", index=False)
    return statistics_df

def plot_competition_figure(competition_df,
                                tf_nuv_col,
                                tf_uv_col,
                                x_ticks_scatter,
                                y_ticks_scatter,
                                y_ticks_uv_bxplt,
                                y_ticks_nuv_bxplt,
                                xlim_scatter,
                                ylim_scatter,
                                ylim_uv_bxplt,
                                ylim_nuv_bxplt,
                                plot_output,
                                palette=PALETTE,
                                repair_col="UVDDB",
                                repair_competition_col="UVDDB_ETS1"):
    competition_df["Color"] = competition_df["Effect_Label"].apply(lambda x: palette[x])
    # Figure parameters and setup
    plt.rcParams["figure.figsize"] = [20,8] # Wide aspect ratio
    # 3 subplots, with the scatterplot being larger by 3:1
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, gridspec_kw={'width_ratios': [3, 2, 2]})
    plt.subplots_adjust(wspace=0.3) # Adjust space between plots 
    # Draw scatterplot
    ax1.scatter(competition_df[repair_col],
                competition_df[repair_competition_col],
                c=competition_df["Color"])
    ax1.axline((1, 1), slope=1, color='black') # Draw the diagonal
    # Draw the boxplots for non-UV and UV signal
    for bxplt_ax, y_axis in ((ax2, tf_uv_col), (ax3, tf_nuv_col)):
        sns.boxplot(x="Effect_Label",
                    y=y_axis,
                    showfliers=False,
                    order=palette.keys(),
                    palette=palette.values(),
                    data=competition_df,
                    saturation=1,
                    ax=bxplt_ax)
    # Adjust tick marks and plot draw limits
    ax1.set_xticks(x_ticks_scatter)
    ax1.set_yticks(y_ticks_scatter)
    ax1.set_yticklabels(y_ticks_scatter, rotation=90, va="center")
    ax2.set_yticks(y_ticks_uv_bxplt)
    ax2.set_yticklabels(y_ticks_uv_bxplt, rotation=90, va="center")
    ax3.set_yticks(y_ticks_nuv_bxplt)
    ax3.set_yticklabels(y_ticks_nuv_bxplt, rotation=90, va="center")
    ax1.set_xlabel('UV-DDB Alone\n (UV-DDB Signal; a.u.)', fontsize=20)
    ax1.set_ylabel('UV-DDB + ETS1\n (UV-DDB Signal; a.u.)', fontsize=20)
    ax2.set_ylabel('ETS1 UV Signal (a.u.)', fontsize=20)
    ax2.set_xlabel('', fontsize=0)
    ax3.set_xlabel('', fontsize=0)
    ax3.set_ylabel('ETS1 Non-UV Signal (a.u.)', fontsize=20)
    ax1.set_xlim(xlim_scatter[0], xlim_scatter[1])
    ax1.set_ylim(ylim_scatter[0], ylim_scatter[1])
    matplotlib.rc('xtick', labelsize=20) 
    matplotlib.rc('ytick', labelsize=20)
    ax2.set_ylim(ylim_uv_bxplt[0], ylim_uv_bxplt[1])
    ax3.set_ylim(ylim_nuv_bxplt[0], ylim_nuv_bxplt[1])
    plt.savefig(fname=plot_output, format='svg')


def competition_group_from_files(group, files=FILES):
    dfs = []
    for experiment, file in FILES:
        df = uac.process_alldata_file(file, aggregate_medians=False, label_has_pydi=False, ln_transform=False)
        df = df[df["Name"].str.startswith(group)].reset_index(drop=True)
        df["Experiment"] = experiment
        dfs.append(df)
    competition_df = pd.concat(dfs).reset_index(drop=True)
    competition_df = competition_df.pivot(index=["Name", "Sequence"], values="Signal", columns="Experiment")
    competition_df = competition_df.reset_index()
    competition_df["Sequence"] = competition_df["Sequence"].apply(lambda x: x[:25])
    competition_df = competition_df.reset_index()
    return competition_df

```

#### (3) Read data; Prepare for Jonckheere in R; Run linregress and save

```python
linregress_output = f"{OUTPUT_FOLDER_TABLE}/Linear_Regression.csv"
lin_out = open(linregress_output, 'w')
lin_out.write("Group,Bin_Method,Experiment,Pvalue\n")
df_dict = {}
for row in meta_data_competition.itertuples():
    df = competition_group_from_files(row.Group)
    df["Competition_Effect"] = df["UVDDB"] - df["UVDDB_ETS1"]
    if row.Method == "by_range":
        bins = bin_by_range(df["Competition_Effect"], row.Percentile)
    elif row.Method == "by_jnb":
        bins = bin_by_jnb(df["Competition_Effect"], row.Percentile)
    df["Effect_Group"] = pd.cut(df["Competition_Effect"],
                                bins,
                                labels=[0, 1, 2])
    df.to_csv(f"{OUTPUT_FOLDER_TABLE}/{row.Group}_{row.Method}.csv", index=False)
    df_dict[(row.Group, row.Method)] = df
    for i in ("ETS1_UV_r1", "ETS1_NUV_r1", "ETS1_UV_r2", "ETS1_NUV_r2"):
        regression = stats.linregress(x=df["Effect_Group"],
                                      y=df[i],
                                      alternative="greater")
        pvalue_scinote = "{:.2E}".format(regression.pvalue)
        lin_out.write(f"{row.Group},{row.Method},{i},{pvalue_scinote}\n")
        print(row.Group, row.Method, i,regression.pvalue, format(regression.slope, ".2f"))
lin_out.close()
```

#### (1) Functions

```python
PRIMER = "CGCACACATACACATACACACATACACATACGCAC"
PDB_GROUPS = ("ELK1",
              "ETS1_monomer",
              "ETS1_dimer", 
              "Creb1", 
              "CTCF", 
              "Myc+Mad",
              "p53_1", 
              "RelA_dimer", 
              "Runx1", 
              "Stat3",
              "TBP",
              "TBP_Ariel_A")


def return_group(x: str):
    """Return a substring representing the group a label belongs to."""
    # Determine how many subgroups deliminated by '_' to return
    n = 3
    if x.startswith("MITF_Ex"):
        n = 4
    elif x.startswith("random_damageable"):
        n = 2
    # Return the subgroup
    return '_'.join(x.split('_')[:n])


def filter_alldata_dataframe(dataframe: pd.DataFrame,
                             primer: str = PRIMER) -> pd.DataFrame:
    """Remove flagged sequence and non-double stranded sequences.

    Given a dataframe with column names in the alldata format, returns
    a dataframe with all flagged rows, rows with no sequence, and
    rows with a sequence without a primer removed.
    """
    alexa = dataframe.columns[7]
    # Filter out flags
    dataframe = dataframe[(dataframe["Cy3Flags"].isnull()) &
                          (dataframe[f"{alexa}Flags"].isnull())]
    # Remove rows with no sequence
    dataframe = dataframe[dataframe["Sequence"].notnull()]
    # Remove rows without primer
    dataframe = dataframe[dataframe["Sequence"].str.endswith(primer)]
    dataframe = dataframe.reset_index(drop=True)
    return dataframe


def replicate_values(dataframe: pd.DataFrame,
                     column: str,
                     min_count: int = 2) -> pd.DataFrame:
    """Return a dataframe of values with a minimum number of replicates.

    Given a dataframe and column, finds the number of replicates of that
    column and returns a dataframe with a minimum amount of replicates.
    """
    # Calculate value counts and return as a dataframe
    count = pd.DataFrame(dataframe[column].value_counts()).reset_index()
    # Rename the columns
    count = count.rename(columns={column: "Count"})
    count = count.rename(columns={"index": column})
    # Query for items that have a minimum value count
    count = count.query("Count >= @min_count")
    return count


def process_alldata_file(file_path: str) -> pd.DataFrame:
    """Given an alldata file, return a dataframe of custom sequences.

    Takes an alldata file and does the following pipeline:
    1. Read as a dataframe
    2. Remove universal sequences
    3. Remove trailing replicate labels so replicate data can be queried
    4. Subset the column into the sequence name, sequence, and signal values

    Returns a dataframe with a Name, Sequence, and Signal columns.
    """
    # Read dataframe
    dataframe = pd.read_csv(file_path, sep='\t')
    alexa = dataframe.columns[7]  # Name is the specific alexa wavelength for the experiment
    dataframe = filter_alldata_dataframe(dataframe)
    # Remove universal UV-Bind Probes
    dataframe = dataframe[~dataframe["Name"].str.startswith("All")]
    dataframe = dataframe.reset_index(drop=True)
    # Remove trailing replicate label for replicate counting
    dataframe["Name_Group"] = dataframe["Name"]\
        .apply(lambda x: '_'.join(x.split('_')[:-1]))
    # Filter for sequences with 2+ replicates
    replicate_counts = replicate_values(dataframe, "Name_Group")
    name_set = set(replicate_counts["Name_Group"])
    dataframe = dataframe[dataframe["Name_Group"]
                          .isin(name_set)].reset_index(drop=True)
    # Subset by name, sequence and signal columns
    dataframe = dataframe[["Name", "Sequence", alexa]]
    dataframe = dataframe.rename(columns={alexa: "Signal"})
    return dataframe


def filter_rep(x):
    """Remove the _r suffix from a label."""
    if re.search(pattern="_r\d*$", string=x):
        return '_'.join(x.split('_')[:-1])
    else:
        return x
    
def decompress_tbp_labels(df):
    dup = df[df["Name"].apply(lambda x: "Ariel_A" in x and "Ariel_G" in x)].copy()
    dup = dup.reset_index(drop=True)
    dup["Group"] = "TBP_Ariel_G"
    df = pd.concat([df, dup]).reset_index(drop=True)
    return df
```

```python
# Read in each file and concatinate into a single df
dfs = []
for experiment, file in FILES:
    df = process_alldata_file(file)
    df["Group"] = df["Name"].apply(lambda x: x.split('-')[0])
    df["Group"] = df["Group"].apply(lambda x: filter_rep(x))
    df = decompress_tbp_labels(df)
    df["Replicate_Number"] = df["Name"].apply(lambda x: int(x.split('_')[-1][1:]))
    df["Name"] = df["Name"].apply(lambda x: filter_rep(x))
    df["Experiment"] = experiment
    dfs.append(df[["Group", "Name","Replicate_Number", "Sequence", "Signal", "Experiment"]])
competition_df = pd.concat(dfs).reset_index(drop=True)

# For ordering the boxplots
median_groups = competition_df.copy()
median_groups = median_groups[median_groups["Experiment"] == "ETS1_NUV_r1"]
median_groups = median_groups[["Group", "Signal"]].groupby(by="Group").aggregate(np.median).reset_index()
median_groups = median_groups[median_groups["Group"].isin(PDB_GROUPS)]
median_groups = median_groups.sort_values(by="Signal", ascending=False).reset_index(drop=True)
median_groups
```
