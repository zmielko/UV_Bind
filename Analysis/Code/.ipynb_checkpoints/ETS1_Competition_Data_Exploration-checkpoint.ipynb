### Imports, global variables, and functions

```python




def return_group(x: str):
    """Return a substring representing the group a label belongs to."""
    # Determine how many subgroups deliminated by '_' to return
    n = 3
    if x.startswith("MITF_Ex"):
        n = 4
    elif x.startswith("random_damageable"):
        n = 2
    # Return the subgroup
    return '_'.join(x.split('_')[:n])


def filter_alldata_dataframe(dataframe: pd.DataFrame,
                             primer: str = PRIMER) -> pd.DataFrame:
    """Remove flagged sequence and non-double stranded sequences.

    Given a dataframe with column names in the alldata format, returns
    a dataframe with all flagged rows, rows with no sequence, and
    rows with a sequence without a primer removed.
    """
    alexa = dataframe.columns[7]
    # Filter out flags
    dataframe = dataframe[(dataframe["Cy3Flags"].isnull()) &
                          (dataframe[f"{alexa}Flags"].isnull())]
    # Remove rows with no sequence
    dataframe = dataframe[dataframe["Sequence"].notnull()]
    # Remove rows without primer
    dataframe = dataframe[dataframe["Sequence"].str.endswith(primer)]
    dataframe = dataframe.reset_index(drop=True)
    return dataframe


def replicate_values(dataframe: pd.DataFrame,
                     column: str,
                     min_count: int = 2) -> pd.DataFrame:
    """Return a dataframe of values with a minimum number of replicates.

    Given a dataframe and column, finds the number of replicates of that
    column and returns a dataframe with a minimum amount of replicates.
    """
    # Calculate value counts and return as a dataframe
    count = pd.DataFrame(dataframe[column].value_counts()).reset_index()
    # Rename the columns
    count = count.rename(columns={column: "Count"})
    count = count.rename(columns={"index": column})
    # Query for items that have a minimum value count
    count = count.query("Count >= @min_count")
    return count


def process_alldata_file(file_path: str) -> pd.DataFrame:
    """Given an alldata file, return a dataframe of custom sequences.

    Takes an alldata file and does the following pipeline:
    1. Read as a dataframe
    2. Remove universal sequences
    3. Remove trailing replicate labels so replicate data can be queried
    4. Subset the column into the sequence name, sequence, and signal values

    Returns a dataframe with a Name, Sequence, and Signal columns.
    """
    # Read dataframe
    dataframe = pd.read_csv(file_path, sep='\t')
    alexa = dataframe.columns[7]  # Name is the specific alexa wavelength for the experiment
    dataframe = filter_alldata_dataframe(dataframe)
    # Remove universal UV-Bind Probes
    dataframe = dataframe[~dataframe["Name"].str.startswith("All")]
    dataframe = dataframe.reset_index(drop=True)
    # Remove trailing replicate label for replicate counting
    dataframe["Name_Group"] = dataframe["Name"]\
        .apply(lambda x: '_'.join(x.split('_')[:-1]))
    # Filter for sequences with 2+ replicates
    replicate_counts = replicate_values(dataframe, "Name_Group")
    name_set = set(replicate_counts["Name_Group"])
    dataframe = dataframe[dataframe["Name_Group"]
                          .isin(name_set)].reset_index(drop=True)
    # Subset by name, sequence and signal columns
    dataframe = dataframe[["Name", "Sequence", alexa]]
    dataframe = dataframe.rename(columns={alexa: "Signal"})
    return dataframe


def filter_rep(x):
    """Remove the _r suffix from a label."""
    if re.search(pattern="_r\d*$", string=x):
        return '_'.join(x.split('_')[:-1])
    else:
        return x
    
def decompress_tbp_labels(df):
    dup = df[df["Name"].apply(lambda x: "Ariel_A" in x and "Ariel_G" in x)].copy()
    dup = dup.reset_index(drop=True)
    dup["Group"] = "TBP_Ariel_G"
    df = pd.concat([df, dup]).reset_index(drop=True)
    return df


def show_scatterplot(plot_df: pd.DataFrame,
                     x: str,
                     y: str,
                     color_on: str,
                     title: str,
                     color_map: str = "viridis"):
    """Show a scatterplot of 2 columns in a dataframe."""
    figure(figsize=(8, 8), dpi=90)
    plt.scatter(plot_df[x],
               plot_df[y],
               c=plot_df[color_on],
               cmap=color_map)
    plt.title(title)
    plt.colorbar()
    plt.axline((1, 1), slope=1, color='black')
    plt.show()


def show_hexbin(plot_df: pd.DataFrame,
                x: str,
                y: str,
                color_on: str,
                title: str,
                color_map: str = "viridis"):
    """Show a hexbin of 2 columns in a dataframe aggregating by median."""
    figure(figsize=(8, 8), dpi=80)
    plt.hexbin(plot_df[x],
               plot_df[y],
               C=plot_df[color_on],
               reduce_C_function=np.median,
               gridsize=50,
               cmap=color_map)
    plt.title(title)
    plt.axline((1, 1), slope=1, color='black')
    plt.colorbar()
    plt.show()
```

### Reading the data

1. Each file in **FILES** is read using uac.process_alldata_file
2. The results are saved as MITF_UVDDB_Competition.csv

```python
# Read in each file and concatinate into a single df
dfs = []
for experiment, file in FILES:
    df = process_alldata_file(file)
    df["Group"] = df["Name"].apply(lambda x: x.split('-')[0])
    df["Group"] = df["Group"].apply(lambda x: filter_rep(x))
    df = decompress_tbp_labels(df)
    df["Replicate_Number"] = df["Name"].apply(lambda x: int(x.split('_')[-1][1:]))
    df["Name"] = df["Name"].apply(lambda x: filter_rep(x))
    df["Experiment"] = experiment
    dfs.append(df[["Group", "Name","Replicate_Number", "Sequence", "Signal", "Experiment"]])
competition_df = pd.concat(dfs).reset_index(drop=True)
# Save table in long format
competition_df.to_csv("ETS1_UVDDB_Competition.csv", index=False)
competition_df
```

#### Binning strategy
There are 2 different binning strategies that have been discussed in some manner or another. Most of them revolve around defining a bin of "weak" as taking the max of datapoints increasing in the competition dataset and flipping it across the diagonal. Then taking the remainder of the datapoints and splitting it in half. 

1. bin_flip_around_diagonal: Does the diagonal flipping method.
2. bin_by_range: Bins in 3 equal groups by range. 

Each has a percentile argument to consider the min and max values as a percentile rather than true min or max to account for outliers (ex/ a 99 percentile choice). 

#### Statistics
Groups are compared using a Wilcoxon rank-sum test with a benjamini-hochberg correction. 

#### Other Parameters
The PALETTE parameter is an ordered dictionary with the name of a group indicated by the keys and the color by the values. 

```python
# This defines the name and color of the 3 bins
PALETTE = OrderedDict([("Weak","#ffeda0"),
                       ("Medium", "#feb24c"),
                       ("Strong", "#f03b20")])


def bin_around_diagonal(binby_1d_data, percentile=100):
    """This uses a """
    min_percentile = 100 - percentile
    min_value = np.percentile(binby_1d_data, min_percentile)
    if min_value >= 0: 
        raise ValueError("Min value below diagonal, cannot perform binning strategy.")
    bin1 = abs(min_value) # Flip aroud the diagonal
    bin2 = ((np.percentile(binby_1d_data, percentile) - bin1) / 2) + bin1 # Take the midpoint of the rest
    return (-math.inf, bin1, bin2, math.inf)


def bin_by_range(binby_1d_data, percentile):
    min_percentile = 100 - percentile
    min_value = np.percentile(binby_1d_data, min_percentile)
    max_value = np.percentile(binby_1d_data, percentile)
    bin_size = (max_value - min_value) / 3
    bin1 = min_value + bin_size
    bin2 = bin1 + bin_size
    return (-math.inf, bin1, bin2, math.inf)


def ranksum_comparison(df, value_column, groups=list(PALETTE.keys())):
    """Perform a Wilcoxon rank sum test with FDR correction."""
    test_results= []
    # For each group, find the wilxocon rank-sum results
    for group_a, group_b in itertools.combinations(groups, 2):
        values_a = df[df["Effect_Group"] == group_a][value_column]
        values_b = df[df["Effect_Group"] == group_b][value_column]
        p_value = stats.ranksums(values_a, values_b).pvalue
        test_results.append((group_a, group_b, p_value))
    result_df = pd.DataFrame(test_results)
    result_df = result_df.rename(columns={0:"Group_A", 1:"Group_B", 2:"P_Value"})
    fdr_values = multipletests(result_df["P_Value"], method="fdr_bh")[1]
    result_df["FDR"] = fdr_values
    return result_df
```

The plot_f5 function does 2 things:

1. Generates a plot in the same display setup as f5, with a scatterplot and 2 boxplots.
2. Runs the ranksum statistics with correction and returns a dataframe of the results. 

```python
def plot_f5(competition_df_wide,
            group,
            non_uv_column,
            uv_column,
            bin_technique,
            percentile = 100,
            plot_output = None):
    # Query the competiton dataframe for data to plot
    plot_df = competition_df_wide.copy()
    plot_df = plot_df[plot_df["Group"] == group]
    if bin_technique == "bin_around_diagonal":
        bins = bin_around_diagonal(plot_df["ETS1_Effect"], percentile)
    elif bin_technique == "bin_by_range":
        bins = bin_by_range(plot_df["ETS1_Effect"], percentile)
    else:
        raise ValueError("Bin technique must be 'bin_around_diagonal' or 'bin_by_range'")
    plot_df["Effect_Group"] = pd.cut(plot_df["ETS1_Effect"],
                                       bins,
                                       labels=list(PALETTE.keys()))
    plot_df["Color"] = plot_df["Effect_Group"].apply(lambda x: PALETTE[x])
    # Plot the data
    # Setup overall plot parameters
    plt.rcParams["figure.figsize"] = [16,9] # Wide aspect ratio
    # 3 subplots, with the scatterplot being larger by 3:1
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, gridspec_kw={'width_ratios': [3, 1, 1]})
    plt.subplots_adjust(wspace=0.3) # Adjust space between plots 
    fig.suptitle(f"Bin technique: '{bin_technique} - Percentile: {percentile}'; Non-UV: {non_uv_column}; UV: {uv_column}") # Title of the figure
    # Scatterplot parameters
    ax1.scatter(plot_df["UVDDB"],
               plot_df["UVDDB_ETS1"],
               c=plot_df["Color"])
    ax1.set_xlabel("UVDDB")
    ax1.set_ylabel("UVDDB_ETS1")
    ax1.axline((1, 1), slope=1, color='black')
    # Boxplot for non UV groups
    sns.boxplot(x="Effect_Group",
                y=non_uv_column,
                showfliers=False,
                order=PALETTE.keys(),
                palette=PALETTE.values(),
                data=plot_df,
                ax = ax2)
    # Boxplot for UV groups
    sns.boxplot(x="Effect_Group",
                y=uv_column,
                showfliers=False,
                order=PALETTE.keys(),
                palette=PALETTE.values(),
                data=plot_df,
                ax = ax3)
    plt.show()
    # If an output locaton is provided, save
    if plot_output:
        plt.savefig(plot_output)
    # Run statistics
    non_uv_statistics_df = ranksum_comparison(plot_df, non_uv_column)
    non_uv_statistics_df["Group"] = group
    non_uv_statistics_df["Column"] = non_uv_column
    uv_statistics_df = ranksum_comparison(plot_df, uv_column)
    uv_statistics_df["Group"] = group
    uv_statistics_df["Column"] = uv_column
    statistics_df = pd.concat([non_uv_statistics_df, uv_statistics_df]).reset_index(drop=True)
    return statistics_df
```
