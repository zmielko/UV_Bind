### ETS1_UVDDB Competition analysis

Data processing, plots, and statistics for the UVDDB vs ETS1 + UVDDB competition assay.

1. Figure 5e

### Files
The following files are used to generate the data:

1. UVDDB_UV_ID24_alldata.txt
2. UVDDBETS1_UV_ID25_alldata.txt
3. ETS1_WC_ID8_alldata.txt
4. ETS1_UV_ID10_alldata.txt

### Overview

UVDDB and ETS1 + UVDDB conditions are compared to one another to examine how the addition of ETS1 might impact UVDDB binding to UV irradiated ETS1 sites. Sequences were binned by the difference between the two conditions into 3 bins. A scatterplot of the comparison with datapoints drawn in a color per bin and a line representing the diagonal is generated. 

To examine what the influence of ETS1 binding specificty might be, ETS1 binding signal was compared for UV irradiated and non-UV conditions for the 3 bins. A wilcoxon rank-sum test and benjamini-hochberg correction for multiple tests was performed with a FDR cuttoff of 0.05 to determine if groups were significantly different or not. Boxplots comparing the groups were generated. 

### 3rd Party Packages

1. Bokeh - Plotting (Scatterplots)
2. Matplotlib - Plotting (Boxplot)
4. Pandas - Dataframe usage
3. Seaborn - Plotting (Boxplots)
4. Scipy - Wilcoxon Rank-Sum test
5. Statsmodels: Multiple test correction

### UV Bind Analysis Core Imports

- uac.process_alldata_file: Used to read aldata.txt files and filter out rows with flagged sequences, 1 replicate, or no sequences. 

** Additional details can be found in the uvbind_analysis_core.py script.

- df: DataFrame

#### Data preprocessing and grouping

1. Read the files for UVDDB, ETS1 + UVDDB, and ETS1 in UV and non-UV conditions.
2. Combine them into a single dataframe
3. Subsect the dataframe for the ETS1 monomer sequence group
4. Bin the dataset into groups based on the ETS1 influence of UVDDB signal

The comparison here uses all sequences individually in non-ln space. 

**Input:** Files given in the FILES global variable

**Output:** Dataframe with the following columns:

1. Name: Label given to the sequence from the design. These include the \_r suffixes as the row represents the individual signals measured.
2. Sequence: The 60bp sequence in the UV-Bind array.
3. ETS1UVDDB: Signal values for the ETS1 + UVDDB binding to UV-irradiated DNA.
4. ETS1_NUV: Signal values for ETS1 binding to non-irradiated DNA.
5. ETS1_UV: Signal values for ETS1 binding to UV-irradiated DNA.
6. UVDDB: Signal values for UVDDB binding to UV-irradiated DNA.
7. UVDDB_Difference: UVDDB - ETS1UVDDB values, representing the change in signal if ETS1 is added along with UVDDB.
8. ETS1_Effect: Bin that the UVDDB_Difference was put into as defined by the BINS and PALETTE global variables.
9. Color: Color used for plotting the data as defined in the PALETTE global variable.

```python
def bins_from_df(df, by=BINBY):
    l = list(df["UVDDB_Difference"].sort_values())
    if by=='range':
        r = l[-1] - l[0]
        s = (r/3)
        b1 = l[0] + s
        b2 = l[0] + s + s
    elif by=="rank":
        s =  math.floor(len(l) / 3)
        b1 = l[s]
        b2 = l[s * 2]
    else:
        raise ValueError("AHHHHHHHHHHHHHHHHHHH")
    return(b1, b2)

# Read files and combine into a single dataframe
binding_data = []
for name, file in FILES:
    df = uac.process_alldata_file(file,
                                  ln_transform=False,
                                  aggregate_medians=True,
                                  label_has_pydi=False)
    df["Dataset"] = name
    binding_data.append(df)
comparison = pd.concat(binding_data).reset_index(drop=True)
# Pivot so that each dataset signal is its own column
comparison = comparison.pivot(index=["Name", "Sequence"],
                              columns='Dataset',
                              values='Signal')
comparison = comparison.reset_index()
comparison = comparison.rename(columns={"Dataset":"Index"})
# Subsection data for the ETS1 monomer sequence group
comparison = comparison.reset_index(drop=True)
# Calculate difference from UVDDB and ETS1UVDDB
#comparison["UVDDB_Difference"] = comparison["UVDDB"] - comparison["ETS1UVDDB"]
comparison["UVDDB_Difference"] = (comparison["UVDDB"] - comparison["ETS1UVDDB"]) / comparison["UVDDB"]
# Bin into groups, add a column for the color
if auto_bin:
    b1, b2 = bins_from_df(comparison)
    BINS = (-math.inf, b1, b2, math.inf)
else:
    BINS = RG_BINS
comparison["ETS1_Effect"] = pd.cut(comparison["UVDDB_Difference"],
                                   BINS,
                                   labels=list(PALETTE.keys()))
comparison["Color"] = comparison["ETS1_Effect"].apply(lambda x: PALETTE[x])
comparison.to_csv(f"{OUTPUT_FOLDER}/{ITEM}_Plotted_Data.csv", index=False)
 
```

```python
for ITEM in LIST:
    NAME_FILTER = ITEM
    OUTPUT_FOLDER = f"Competition_Validation/bin_by_rank/{ITEM}"
    os.makedirs(OUTPUT_FOLDER, exist_ok=True)
    def bins_from_df(df, by=BINBY):
        l = list(df["UVDDB_Difference"].sort_values())
        if by=='range':
            r = l[-1] - l[0]
            s = (r/3)
            b1 = l[0] + s
            b2 = l[0] + s + s
        elif by=="rank":
            s =  math.floor(len(l) / 3)
            b1 = l[s]
            b2 = l[s * 2]
        else:
            raise ValueError("AHHHHHHHHHHHHHHHHHHH")
        return(b1, b2)

    # Read files and combine into a single dataframe
    binding_data = []
    for name, file in FILES:
        df = uac.process_alldata_file(file,
                                      ln_transform=False,
                                      aggregate_medians=False,
                                      label_has_pydi=False)
        df["Dataset"] = name
        binding_data.append(df)
    comparison = pd.concat(binding_data).reset_index(drop=True)
    # Pivot so that each dataset signal is its own column
    comparison = comparison.pivot(index=["Name", "Sequence"],
                                  columns='Dataset',
                                  values='Signal')
    comparison = comparison.reset_index()
    comparison = comparison.rename(columns={"Dataset":"Index"})
    # Subsection data for the ETS1 monomer sequence group
    comparison = comparison[comparison["Name"].str.startswith(NAME_FILTER)]
    comparison = comparison.reset_index(drop=True)
    # Calculate difference from UVDDB and ETS1UVDDB
    comparison["UVDDB_Difference"] = comparison["UVDDB"] - comparison["ETS1UVDDB"]
    # Bin into groups, add a column for the color
    if auto_bin:
        b1, b2 = bins_from_df(comparison)
        BINS = (-math.inf, b1, b2, math.inf)
    else:
        BINS = RG_BINS
    comparison["ETS1_Effect"] = pd.cut(comparison["UVDDB_Difference"],
                                       BINS,
                                       labels=list(PALETTE.keys()))
    comparison["Color"] = comparison["ETS1_Effect"].apply(lambda x: PALETTE[x])
    comparison.to_csv(f"{OUTPUT_FOLDER}/{ITEM}_Plotted_Data.csv", index=False)
    # Calculate plot ranges
    x_range = plot_range_1d(comparison["UVDDB"], 0.05)
    y_range = plot_range_1d(comparison["ETS1UVDDB"], 0.05)
    # Create coordiantes to draw the diagonal line
    max_val = max([max(x_range), max(y_range)])
    min_val = min([min(x_range), min(y_range)])
    diagonal = (min_val, max_val) # (x, y), (x, y) will return the diagonal
    plot = scatterplot(comparison,
                "UVDDB",
                "ETS1UVDDB",
                x_range,
                y_range,
                diagonal,
                font_size='10pt' )
    plot.output_backend = 'svg'
    export_svg(plot, filename=f"{OUTPUT_FOLDER}/{ITEM}_UVDDB_ETS1UVDDB_Scatterplot.svg")
    # Convert from wide to long 
    comparison_long = pd.melt(comparison,
                              id_vars=["Name", "Sequence", "ETS1_Effect", "Color"],
                              value_vars=list(map(lambda x: x[0], FILES)))
    condition_results = []
    for condition in ("ETS1_NUV", "ETS1_UV"):
        # Query for the condition
        df = comparison_long[comparison_long["Dataset"] == condition]
        df = df.reset_index(drop=True)
        test_results= []
        # For each group, find the wilxocon rank-sum results
        for group_a, group_b in itertools.combinations(("Weak", "Medium", "Strong"),
                                                        2):
            values_a = df[df["ETS1_Effect"] == group_a]["value"]
            values_b = df[df["ETS1_Effect"] == group_b]["value"]
            p_value = stats.ranksums(values_a, values_b).pvalue
            print(values_a, values_b)
            test_results.append((group_a, group_b, p_value))
        # Store the results as a dataframe
        wilcoxon_df = pd.DataFrame(test_results)
        wilcoxon_df = wilcoxon_df.rename(columns={0:"Group_A",
                                                  1:"Group_B",
                                                  2:"P_Value"})
        # Calculate the false discovery rates and add to the dataframe
        fdr_values = multipletests(wilcoxon_df["P_Value"], method="fdr_bh")[1]
        wilcoxon_df["FDR"] = fdr_values
        wilcoxon_df["Condition"] = condition
        condition_results.append(wilcoxon_df)
    # Combine the results for both conditions and save as one table
    condition_wilcoxon_tests = pd.concat(condition_results)
    condition_wilcoxon_tests.to_csv(f"{OUTPUT_FOLDER}/{ITEM}_Panel_E_Boxplot_Statistics.csv", 
                                    index=False)
    # Boxplots
    plt.rcParams["figure.figsize"] = (5,10) # Draw in a tall rectangle aspect ratio
    for condition in ("ETS1_NUV", "ETS1_UV"):
        sns.boxplot(x="ETS1_Effect",
                    y="value",
                    data=comparison_long.query(f"Dataset == @condition"),
                    palette=PALETTE.values(),
                    order=PALETTE.keys(),
                    showfliers=False)
        plt.savefig(fname=f"{OUTPUT_FOLDER}/{ITEM}_Panel_E_Boxplot_{condition}.svg",
                    format='svg')
        plt.close()
```

#### Scatterplot of UVDDB vs ETS1 + UVDDB

Define plot settings for bokeh and draw a scatterplot of the comparison.

**Input:** comparison dataframe (output of Data preprocessing and grouping)

**Output:** A scatterplot comparing UVDDB on the x-axis with ETS1 + UVDDB on the y-axis with a line representing the diagonal and colors representing each group as defined in the *Color* column.

```python
def scatterplot(dataframe: pd.DataFrame,
                     column_a: str,
                     column_b: str,
                     x_range: tuple,
                     y_range,
                     diag,
                     font: str = 'ariel',
                     font_size: str = '0pt',
                     circle_size: int = 10,
                     line_width: int = 2,
                     outline_width: int = 2,
                     outline_color: str = "black"):
    """Return a bokeh figure object of a scatterplot.

    Given a dataframe, two columns of data to plot as column a
    and b, a tuple of two values for the plot min and max values,
    and settings for font, font_size, circle_size, and color, 
    returns a bokeh figure object of a scatterplot. 

    The plot contains a black border with the diagonal drawn. 
    """
    source = ColumnDataSource(dataframe)
    p = figure(x_range=x_range, y_range=y_range)
    # Draw the data points
    p.circle(column_a,
             column_b,
             color="Color",
             size=circle_size,
             source=source)
    # Draw a diagonal line
    p.line(diag, diag, color="black", width=line_width)
    # Settings for label text
    p.xaxis.major_label_text_font_size = font_size
    p.yaxis.major_label_text_font_size = font_size
    p.xaxis.major_label_text_font = font
    p.yaxis.major_label_text_font = font
    p.xaxis.minor_tick_line_color = None
    p.yaxis.minor_tick_line_color = None 
    # Settings for the border
    p.outline_line_width = outline_width
    p.outline_line_color = outline_color
    # Remove toolbar and grids
    p.toolbar_location = None
    p.xgrid.visible = False
    p.ygrid.visible = False
    return p

def plot_range_1d(x: list[int], extension: float):
    """Define a plot range for a 1-D dataset."""
    x_range = max(x) - min(x)
    lower = min(x) - (x_range * extension)
    upper = max(x) + (x_range * extension)
    return (lower, upper)


# Calculate plot ranges
x_range = plot_range_1d(comparison["UVDDB"], 0.05)
y_range = plot_range_1d(comparison["ETS1UVDDB"], 0.05)
# Create coordiantes to draw the diagonal line
max_val = max([max(x_range), max(y_range)])
min_val = min([min(x_range), min(y_range)])
diagonal = (min_val, max_val) # (x, y), (x, y) will return the diagonal
plot = scatterplot(comparison,
            "UVDDB",
            "ETS1UVDDB",
            x_range,
            y_range,
            diagonal,
            font_size='10pt' )
show(plot)
#plot.output_backend = 'svg'
#export_svg(plot, filename=f"{OUTPUT_FOLDER}/UVDDB_ETS1UVDDB_Scatterplot.svg")
```

### ETS1 Signal Compared to ETS1 Effect on UVDDB Signal

#### Statistical Analysis

1. Arrange the dataframe into a long format
2. Calculate the wilcoxon rank-sum test comparison for each group
3. Determine the false discover rate for the comparisons using a benjamini-hochberg test for multiple corrections.
4. Save the results

\* Long format is done for convienence in using seaborn/matplotlib.

**Input:** comparison dataframe (output of Data preprocessing and grouping)

**Output:** Dataframe with the following columns:

1. Group_A: A group representing either a "weak", "medium" or "strong" effect. 
2. Group_B: The group that Group_A is being compared to.
3. P_Value: P-value as calculated by a wilcoxon rank-sum test.
4. FDR: False discovery rate as calcualted by a benjamini-hochberg correction for multiple tests. The FDR is based on each condition (All tests in non-UV ETS1, all tests in UV ETS1) individually.

```python
# Convert from wide to long 
comparison_long = pd.melt(comparison,
                          id_vars=["Name", "Sequence", "ETS1_Effect", "Color"],
                          value_vars=list(map(lambda x: x[0], FILES)))
condition_results = []
for condition in ("ETS1_NUV", "ETS1_UV"):
    # Query for the condition
    df = comparison_long[comparison_long["Dataset"] == condition]
    df = df.reset_index(drop=True)
    test_results= []
    # For each group, find the wilxocon rank-sum results
    for group_a, group_b in itertools.combinations(("Weak", "Medium", "Strong"),
                                                    2):
        values_a = df[df["ETS1_Effect"] == group_a]["value"]
        values_b = df[df["ETS1_Effect"] == group_b]["value"]
        p_value = stats.ranksums(values_a, values_b).pvalue
        print(values_a, values_b)
        test_results.append((group_a, group_b, p_value))
    # Store the results as a dataframe
    wilcoxon_df = pd.DataFrame(test_results)
    wilcoxon_df = wilcoxon_df.rename(columns={0:"Group_A",
                                              1:"Group_B",
                                              2:"P_Value"})
    # Calculate the false discovery rates and add to the dataframe
    fdr_values = multipletests(wilcoxon_df["P_Value"], method="fdr_bh")[1]
    wilcoxon_df["FDR"] = fdr_values
    wilcoxon_df["Condition"] = condition
    condition_results.append(wilcoxon_df)
# Combine the results for both conditions and save as one table
condition_wilcoxon_tests = pd.concat(condition_results)
condition_wilcoxon_tests.to_csv(f"{OUTPUT_FOLDER}/Panel_E_Boxplot_Statistics.csv", 
                                index=False)
```

```python
def range_bin(iterable):
    v_max = max(iterable)
    v_min = min(iterable)
    r = v_max - v_min
    s = (r/3)
    b1 = v_min + s
    b2 = v_min + s + s
    return (b1, b2)

binding_data = []
for name, file in FILES:
    df = uac.process_alldata_file(file,
                                  ln_transform=False,
                                  aggregate_medians=False,
                                  label_has_pydi=False)
    df["Dataset"] = name
    binding_data.append(df)
comparison = pd.concat(binding_data).reset_index(drop=True)
# Pivot so that each dataset signal is its own column
comparison = comparison.pivot(index=["Name", "Sequence"],
                              columns='Dataset',
                              values='Signal')
comparison = comparison.reset_index()
comparison = comparison.rename(columns={"Dataset":"Index"})
# Subsection data for the ETS1 monomer sequence group
#comparison = comparison[comparison["Name"].str.startswith(NAME_FILTER)]
comparison = comparison.reset_index(drop=True)
# Calculate difference from UVDDB and ETS1UVDDB
comparison["UVDDB_Difference"] = comparison["UVDDB"] - comparison["ETS1UVDDB"]
comparison = comparison.reset_index(drop=True)
# Bin into groups, add a column for the color
b1, b2 = range_bin(comparison["ETS1_UV"])
BINS = (-math.inf, b1, b2, math.inf)
comparison["ETS1_Bins"] = pd.cut(comparison["ETS1_UV"],
                                   BINS,
                                   labels=list(PALETTE.keys()))
comparison["Color"] = comparison["ETS1_Bins"].apply(lambda x: PALETTE[x])
comparison.to_csv(f"{OUTPUT_FOLDER}/By_ETS1_NUV_Strength_Scatterplot.csv", index=False)
print(BINS)
```

```python
# Convert from wide to long 
comparison_long = pd.melt(comparison,
                          id_vars=["Name", "Sequence", "ETS1_Bins", "Color"],
                          #value_vars=list(map(lambda x: x[0], FILES)))
                          value_vars="UVDDB_Difference")
condition_results = []
# Query for the condition
df = comparison_long
df = df.reset_index(drop=True)
test_results= []
# For each group, find the wilxocon rank-sum results
for group_a, group_b in itertools.combinations(("Weak", "Medium", "Strong"),
                                                2):
    values_a = df[df["ETS1_Bins"] == group_a]["value"]
    values_b = df[df["ETS1_Bins"] == group_b]["value"]
    p_value = stats.ranksums(values_a, values_b).pvalue
    test_results.append((group_a, group_b, p_value))
# Store the results as a dataframe
wilcoxon_df = pd.DataFrame(test_results)
wilcoxon_df = wilcoxon_df.rename(columns={0:"Group_A",
                                          1:"Group_B",
                                          2:"P_Value"})
# Calculate the false discovery rates and add to the dataframe
fdr_values = multipletests(wilcoxon_df["P_Value"], method="fdr_bh")[1]
wilcoxon_df["FDR"] = fdr_values
condition_results.append(wilcoxon_df)
# Combine the results for both conditions and save as one table
condition_wilcoxon_tests = pd.concat(condition_results)
#condition_wilcoxon_tests.to_csv(f"{OUTPUT_FOLDER}/Panel_E_Boxplot_Statistics.csv", 
#                                index=False)
condition_wilcoxon_tests
```

```python
def scatterplot(dataframe: pd.DataFrame,
                     column_a: str,
                     column_b: str,
                     x_range: tuple,
                     y_range,
                     diag,
                     font: str = 'ariel',
                     font_size: str = '0pt',
                     circle_size: int = 10,
                     line_width: int = 2,
                     outline_width: int = 2,
                     outline_color: str = "black"):
    """Return a bokeh figure object of a scatterplot.

    Given a dataframe, two columns of data to plot as column a
    and b, a tuple of two values for the plot min and max values,
    and settings for font, font_size, circle_size, and color, 
    returns a bokeh figure object of a scatterplot. 

    The plot contains a black border with the diagonal drawn. 
    """
    source = ColumnDataSource(dataframe)
    p = figure(x_range=x_range, y_range=y_range)
    # Draw the data points
    p.circle(column_a,
             column_b,
             color="Color",
             size=circle_size,
             source=source)
    # Draw a diagonal line
    p.line(diag, diag, color="black", width=line_width)
    # Settings for label text
    p.xaxis.major_label_text_font_size = font_size
    p.yaxis.major_label_text_font_size = font_size
    p.xaxis.major_label_text_font = font
    p.yaxis.major_label_text_font = font
    p.xaxis.minor_tick_line_color = None
    p.yaxis.minor_tick_line_color = None 
    p.xgrid.visible = False
    p.ygrid.visible = False
    return p


# Calculate plot ranges
x_range = plot_range_1d(comparison["UVDDB"], 0.05)
y_range = plot_range_1d(comparison["ETS1UVDDB"], 0.05)
# Create coordiantes to draw the diagonal line
max_val = max([max(x_range), max(y_range)])
min_val = min([min(x_range), min(y_range)])
diagonal = (min_val, max_val) # (x, y), (x, y) will return the diagonal
plot = scatterplot(comparison,
            "UVDDB",
            "ETS1UVDDB",
            x_range,
            y_range,
            diagonal,
            font_size='10pt' )
plot.output_backend = 'svg'
show(plot)
```
