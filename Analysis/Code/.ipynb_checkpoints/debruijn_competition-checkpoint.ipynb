```python
from collections import OrderedDict
import itertools
import math
import os

from bokeh.io import export_svg
from bokeh.models import ColumnDataSource
from bokeh.plotting import figure, show
import matplotlib.pyplot as plt
from matplotlib import rcParams
import pandas as pd
import numpy as np
from scipy import stats
import seaborn as sns
from statsmodels.stats.multitest import multipletests


import uvbind_analysis_core as uac

ADJ = False
NBINS = 6
BBY = "rank"





ALLDATA_FOLDER = "../../Data/AllData_Files"
gpr_meta = pd.read_csv("../../Data/GPR_Meta_Data.txt", sep='\t')
pdb_design = gpr_meta[gpr_meta["ArrayID"] == 85438].reset_index(drop=True)
files = []
for row in pdb_design.itertuples():
    name = f"{row.Protein}_{row.Condition}_{row.ID}"
    file = f"{ALLDATA_FOLDER}/{name}_alldata.txt"
    files.append((name, file))
files
```

```python
def read_debruijn(file_path: str, adj=False) -> pd.DataFrame:
    # Read dataframe
    dataframe = pd.read_csv(file_path, sep='\t')
    if adj:
        alexa = dataframe.columns[-1]
    else:
        alexa = dataframe.columns[7]
    dataframe = uac.filter_alldata_dataframe(dataframe)
    # Remove universal UV-Bind Probes
    dataframe = dataframe[dataframe["Name"].str.startswith("All_9mer")]
    dataframe = dataframe.reset_index(drop=True)
    # Subset by name, sequence and signal columns
    dataframe = dataframe[["Name", "Sequence", alexa]]
    dataframe = dataframe.rename(columns={alexa: "Signal"})
    return dataframe
dfs = []
for name, file in files:
    df = read_debruijn(file, adj=ADJ)
    df["Dataset"] = name
    dfs.append(df)
comparison = pd.concat(dfs).reset_index(drop=True)
# Pivot so that each dataset signal is its own column
comparison = comparison.pivot(index=["Name", "Sequence"],
                              columns='Dataset',
                              values='Signal')
comparison = comparison.reset_index()
comparison = comparison.rename(columns={"Dataset":"Index"})
comparison["UVDDB_Difference"] = comparison["UVDDB_UV_ID24"] - comparison["UVDDBETS1_UV_ID25"]
#comparison.to_csv("Competition_Val/Debruijn_Data_85438.csv", index=False)

plot_range = uac.plot_range_from_x_y(comparison["UVDDB_UV_ID24"],
                                     comparison["UVDDBETS1_UV_ID25"])
p = figure(x_range=plot_range, y_range=plot_range)
p.circle(comparison["UVDDB_UV_ID24"],
         comparison["UVDDBETS1_UV_ID25"],
         color='lightblue')
p.line(plot_range, plot_range, color="black", width=1)
show(p)
```

```python
#### BIN BY AROUND LINE
effect = comparison["ETS1_Effect"].copy()
above_diag_effect = sorted(list(filter(lambda x: x < 0, effect)))
above_diag_effect
IQR = np.quantile(above_diag_effect, 0.75) - np.quantile(above_diag_effect, 0.25)
outlier = np.quantile(above_diag_effect, 0.75) - (IQR * 1.5)
outlier
bin_size = outlier * -2
bin_size
nb = outlier
bins = [-np.inf]
while nb < max(effect):
    bins.append(nb)
    nb += bin_size
bins.append(np.inf)
bins
```

```python
############ BIN BY RANGE
def bins_from_df(df, by, nbin=NBINS):
    l = list(df["UVDDB_Difference"].sort_values())
    if by=='range':
        ed = np.quantile(l, 0.999)
        st = np.quantile(l, 0.001)
        r =  ed - st
        s = (r/nbin)
        binlist = []
        newbin = s
        for i in range(nbin - 1):
            binlist.append(newbin)
            newbin += s
    elif by=="rank":
        s =  math.floor(len(l) / nbin)
        binlist = []
        newbin = s
        for i in range(nbin - 1):
            binlist.append(l[newbin])
            newbin += s        
    return binlist
blist = bins_from_df(comparison, by=BBY)
blist.insert(0, -np.inf)
blist.append(np.inf)
BINS = blist
PALETTE = Category10[len(BINS) - 1]
ORDER = list(range(len(BINS) - 1))

comparison["ETS1_Effect"] = pd.cut(comparison["UVDDB_Difference"],
                                   BINS, 
                                   labels=ORDER)
comparison["Color"] = comparison["ETS1_Effect"].apply(lambda x: PALETTE[x])
comparison
```

```python
plot_range = uac.plot_range_from_x_y(comparison["UVDDB_UV_ID24"],
                                     comparison["UVDDBETS1_UV_ID25"])
p = figure(x_range=plot_range, y_range=plot_range)
p.circle(comparison["UVDDB_UV_ID24"],
         comparison["UVDDBETS1_UV_ID25"],
         color=comparison['Color'])
p.line(plot_range, plot_range, color="black", width=1)
show(p)
```

```python
ORDER = list(map(str, range(len(BINS) - 1)))
comparison["ETS1_Effect"] = comparison["ETS1_Effect"].apply(lambda x: str(x))
def bxplt(name_nuv, name_uv, comparison, ORDER=ORDER):
    pairs = list(itertools.combinations(ORDER, 2))
    plt.rcParams["figure.figsize"] = (10, 5) # Draw in a tall rectangle aspect ratio
    f, axs = plt.subplots(1, 2)
    sns.boxplot(x="ETS1_Effect",
                y=name_nuv,
                data=comparison,
                order=ORDER,
                showfliers=False,
                ax = axs[0])
    sns.boxplot(x="ETS1_Effect",
                y=name_uv,
                data=comparison,
                order=ORDER,
                showfliers=False,
                ax=axs[1])
    annotator_nuv = Annotator(axs[0],
                              data=comparison,
                              x="ETS1_Effect",
                              y=name_nuv,
                              order=ORDER,
                             pairs=pairs)
    annotator_nuv.configure(test="Mann-Whitney",
                            text_format='star',
                            loc='outside',
                            comparisons_correction="fdr_bh")
    annotator_nuv.apply_and_annotate()
    annotator_uv = Annotator(axs[1],
                              data=comparison,
                              x="ETS1_Effect",
                              y=name_uv,
                              order=ORDER,
                             pairs=pairs)
    annotator_uv.configure(test="Mann-Whitney",
                            text_format='star',
                            loc='outside',
                            comparisons_correction="fdr_bh")
    annotator_uv.apply_and_annotate()
```
