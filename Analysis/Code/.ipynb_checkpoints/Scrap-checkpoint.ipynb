```python
def kl_divergence_list(pwm_a, pwm_b) -> list[float]:
    """Returns the sum of divergences"""
    kl_list = []
    for i in range(len(pwm_a)):
        np_matrix_a = np.array(list(pwm_a.T[i]))
        np_matrix_b = np.array(list(pwm_b.T[i]))
        kl_list.append(kl_divergence(np_matrix_a, np_matrix_b))
    return kl_list

def matrix_dictionary_from_meta_data(meta_dataframe: pd.DataFrame,
                                     matrix_name: str,
                                     trim_matrix=True) -> dict:
    """Given a meta data file, reads all matricies from a given name

    Given the name of a type of matrix (Enrichment, Probability, etc),
    reads all matricies as defined in the meta data file and returns a 
    dictionary where the keys are the experiment ID and values are the
    matricies in long form.
    """
    matrix_dict = {}
    for row in meta_dataframe.itertuples():
        prefix = f"{DATA_FOLDER}/{row.Protein}_{row.Condition}_ID{row.ID}"
        if row.File == "Default":
            file_path = f"{prefix}_7mers_pwm.txt"
        else:
            file_path = f"{prefix}_pwm_from_top_canonical.txt"
        pwm = read_pwm_matrix(file_path, matrix_name)
        if row.Reverse_Complement == 1:
            pwm = reverse_complement_matrix(pwm)
        if trim_matrix:
            # Trim PWM
            pwm = pwm.iloc[row.Start_Inclusive:row.End_Exclusive, :]
        pwm = pwm.reset_index(drop=True)
        matrix_dict[row.ID] = pwm
    return matrix_dict
```

```python


def kmer_scatterplot(dataframe: pd.DataFrame,
                     column_a: str,
                     column_b: str,
                     plot_range: tuple,
                     ticker_list: list,
                     font: str = 'ariel',
                     font_size: str = '0pt',
                     circle_size: int = 10,
                     line_width: int = 2,
                     outline_width: int = 2,
                     outline_color: str = "black"):
    """Return a bokeh figure object of a scatterplot.

    Given a dataframe, two columns of data to plot as column a
    and b, a tuple of two values for the plot min and max values,
    and settings for font, font_size, circle_size, and color, 
    returns a bokeh figure object of a scatterplot. 

    The plot contains a black border with the diagonal drawn. 
    """
    source = ColumnDataSource(dataframe)
    p = figure(x_range=plot_range, y_range=plot_range)
    # Draw the data points
    p.circle(column_a,
             column_b,
             color="Color",
             size=circle_size,
             source=source)
    # Draw a diagonal line
    p.line(plot_range, plot_range, color="black", width=line_width)
    # Settings for label text
    p.xaxis.major_label_text_font_size = font_size
    p.yaxis.major_label_text_font_size = font_size
    p.xaxis.major_label_text_font = font
    p.yaxis.major_label_text_font = font
    # Settings for the ticker and tick marks
    p.xaxis.ticker = ticker_list
    p.yaxis.ticker = ticker_list
    # Settings for the border
    p.outline_line_width = outline_width
    p.outline_line_color = outline_color
    # Remove toolbar and grids
    p.toolbar_location = None
    p.xgrid.visible = False
    p.ygrid.visible = False
    return p
```

```python
def plot_df_from_var_df2(tf, var_id, yticker, output):
    t = variation_dict[tf].copy()
    t = t[t["Variant_ID"] == var_id]
    # Remove WT 
    wt = t[t["Position"] < 0]
    wt_len = len(wt)
    if wt_len != 1:
        raise ValueError("WT Search found {wt_len} entries.")
    wt_nuv = float(wt.Signal_Non_UV)
    wt_uv = float(wt.Signal_UV)
    var = t[t["Position"] > 0].reset_index(drop=True)
    # Calculate relative plot positions 
    min_pos = min(var["Position"])
    nuv_plot_positions = []
    for row in var.itertuples():
        # If single mutant
        if len(row.Variant) == 1:
            pos = row.Position - min_pos + 1
        # If double mutant
        elif len(row.Variant) == 2:
            pos = row.Position - min_pos + 1.5
        else:
            raise ValueError("Variant length not 1 or 2.")
        nuv_plot_positions.append(pos)
    var["Plot_Positions_NUV"] = nuv_plot_positions
    max_pos = max(nuv_plot_positions)
    offset = max_pos + 2
    var["Plot_Positions_UV"] = var["Plot_Positions_NUV"] + offset - 1
    #########
    colors = []
    for row in var.itertuples():
        if row.Signal_UV > row.Prediction_Upper:
            colors.append('#8db0fe')
        elif row.Signal_UV < row.Prediction_Lower:
            colors.append('#f4987a')
        else:
            colors.append("grey")
    var["Color"] = colors
    # Convert to a long format df
    nuv_data = var[["Signal_Non_UV", "Is_Significant", "Variant", "Plot_Positions_NUV", "Color"]]
    nuv_data = nuv_data.rename(columns={"Signal_Non_UV":"Signal",
                                        "Plot_Positions_NUV":"Plot_Position"})
    nuv_data["Group"] = "NUV"
    uv_data = var[["Signal_UV", "Is_Significant", "Variant", "Plot_Positions_UV", "Color"]]
    uv_data = uv_data.rename(columns={"Signal_UV":"Signal", "Plot_Positions_UV":"Plot_Position"})
    uv_data["Group"] = "UV"
    plot_df = pd.concat([nuv_data, uv_data]).reset_index(drop=True)
    max_plot_pos = max(plot_df["Plot_Position"])
    circle_size=5
    plot_y_offset = (max(plot_df["Signal"]) - min(plot_df["Signal"])) * 0.1
    y_range = (min(plot_df["Signal"]) - plot_y_offset, max(plot_df["Signal"]) + plot_y_offset)
    x_range = (0, max_plot_pos + 1)
    
    return plot_df


plot_df_from_var_df2("EGR1", 6, (np.log(2500), np.log(10000), np.log(40000)), f"{OUTPUT_MAIN}/Figure_3e.svg")
```

```python
def highlight_plot(df: pd.DataFrame,
                   column_x: str,
                   column_y: str,
                   hue: str,
                   palette: tuple[str, str],
                   order: tuple[str, str],
                   x_ticks: tuple[int, int],
                   output: str,
                   ax,
                  x_limit):
    sns.set_style('ticks')
    #L=5.8
    #U=13
    p = sns.stripplot(data=df,
                      x=column_x,
                      y=column_y,
                      hue=hue,
                      order=order,
                      orient='h',
                      jitter=False,
                      s=15,
                      palette=['blue', 'red'],
                      alpha=0.5,
                      linewidth=0,
                      ax=ax)
    ax.get_legend().remove()
    p.set(xlabel=None)  # remove the axis label   
    p.set(ylabel=None)  # remove the axis label  
    ax.set_xlim(x_limit[0], x_limit[1])
    ax.set_xticks([])
    ax.set_yticks([])


def kde_plot(df,
             column_x,
             hue,
             output,
             x_limit,
             y_limit,
             ax,
             show_x_ticks=False):
    p = sns.kdeplot(data=df,
                    x=column_x,
                    hue=hue,
                    fill=True,
                    common_norm=False,
                    palette=["blue", 'red'],
                    alpha=0.5,
                    linewidth=8,
                    ax=ax)
    ax.set_xlim(x_limit[0], x_limit[1])
    ax.set_ylim(y_limit[0], y_limit[1])
    ax.get_legend().remove()
    plt.legend('',frameon=False)
    p.set(xlabel=None)  # remove the axis label   
    p.set(ylabel=None)  # remove the axis label    
    p.yaxis.major_tick_line_color = None
    p.yaxis.minor_tick_line_color = None
    if show_x_ticks:
        p.set(xticks=[np.log(1000), np.log(10000), np.log(100000)])
        p.set(xticklabels=['1000', '10000', '100000'])
    else:
        ax.set_xticks([])
    ax.set_yticks([])
    return p

groups = ("Concat","SNV", "Consensus")
for protein, highlights in (("CREB1", CREB1_HIGHLIGHT),
                            ("EGR1", EGR1_HIGHLIGHT)):
    fig, ax_array = plt.subplots(4)
    fig.set_size_inches(10, 10)
    kde_x, kde_y = KDE_BOUNDS[protein]
    plot_df = relabeled_dfs[protein].copy()
    plot_df = plot_df[plot_df["Name"].isin(set(highlights))]
    plot_df = pd.melt(plot_df,
                      id_vars=["Name"],
                      value_vars=["Signal_Non_UV", "Signal_UV"])
    highlight_plot(plot_df,
                   "value",
                   "Name",
                   "variable",
                   ('blue', 'red'),
                   highlights,
                   [np.log(1000), np.log(10000), np.log(100000)],
                  f"{OUTPUT_FOLDER_MAIN}/{protein}_Highlight_Plot.svg",
                  ax_array[0],
                  kde_x)
    protein_df = relabeled_dfs[protein].copy()
    protein_df = pd.melt(protein_df, id_vars=["Name", "Group"], value_vars=["Signal_Non_UV", "Signal_UV"])
    for idx, group in enumerate(groups):
        plot_df = protein_df[protein_df["Group"] == group]
        plot_x_ticks = group == "Consensus"
        kde_plot(plot_df,
                 "value",
                 "variable", 
                 f"{OUTPUT_FOLDER_MAIN}/{protein}_{group}_kde_plot.svg",
                 kde_x,
                 kde_y,
                 ax_array[idx+1],
                 plot_x_ticks)
    plt.subplots_adjust(hspace=0.0000001)
    plt.savefig(fname=f"{OUTPUT_FOLDER_MAIN}/{protein}_KDE.svg",format='svg')
    plt.close()

        
```

```python
from collections import namedtuple

Regression_Report = namedtuple("Regression_Report", ["Condition_X",
                                                     "Condition_Y",
                                                     "Prescale_R2",
                                                     "Prescale_Slope",
                                                     "Prescale_Slope_CI",
                                                     "Prescale_Intercept",
                                                     "Prescale_Intercept_CI",
                                                     "Postscale_R2",
                                                     "Postscale_Slope",
                                                     "Postscale_Slope_CI",
                                                     "Postscale_Intercept",
                                                     "Postscale_Intercept_CI"])


def generate_regression_report(prescale_regression,
                               postscale_regression,
                               len_values,
                               column_x,
                               column_y,
                               ci_percent=0.95):
    tinv = lambda p, df: abs(stats.t.ppf(p/2, df))
    ts = tinv(1 - ci_percent, len_values-2)
    return  Regression_Report(Condition_X=column_x,
                             Condition_Y=column_y,
                             Prescale_R2=prescale_regression.rvalue ** 2,
                             Prescale_Slope=prescale_regression.slope,
                             Prescale_Slope_CI=prescale_regression.stderr * ts,
                             Prescale_Intercept=prescale_regression.intercept,
                             Prescale_Intercept_CI=prescale_regression.intercept_stderr * ts,
                             Postscale_R2=postscale_regression.rvalue ** 2,
                             Postscale_Slope=postscale_regression.slope,
                             Postscale_Slope_CI=postscale_regression.stderr * ts,
                             Postscale_Intercept=postscale_regression.intercept,
                             Postscale_Intercept_CI=postscale_regression.intercept_stderr * ts)


def scale_column(dataframe: pd.DataFrame,
                 column_name: str,
                 slope: float,
                 intercept: float) -> pd.DataFrame:
    """Scale a column in a dataframe.

    Given a dataframe, column_name, slope, and intercept, scales a column's
    values such that the slope is equal to 1.
    """
    # Apply the transformation
    dataframe[f"{column_name}_Scaled"] = dataframe[column_name]\
        .apply(lambda x: (x - intercept) / (slope))
    dataframe = dataframe.reset_index(drop=True)
    return dataframe

def scale_on_no_pydi(dataframe, nouv_column, uv_column):
    no_pydi_df = dataframe[~dataframe["Has_PyDi"]]
    prescale_regression = scipy.stats.linregress(no_pydi_df[nouv_column],
                                                  no_pydi_df[uv_column])
    dataframe[f"{uv_column}_Scaled"] = dataframe[uv_column]\
        .apply(lambda x: (x - prescale_regression.intercept) / (prescale_regression.slope))
    post_no_pydi_df = dataframe[~dataframe["Has_PyDi"]]
    postscale_regression = scipy.stats.linregress(post_no_pydi_df[nouv_column],
                                                   post_no_pydi_df[f"{uv_column}_Scaled"])  
    regression_report = generate_regression_report(prescale_regression,
                                                   postscale_regression,
                                                   len(no_pydi_df),
                                                   nouv_column,
                                                   uv_column)
    return (dataframe, regression_report)

```
