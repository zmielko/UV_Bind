### Supplementary Figure 2

This analysis covers replicate comparisons of ETS1 for E-score, and Z-score k-mer data as well as custom probes. This includes calcualting R2 values and creating scatterplots.

#### FS4b,c: K-mer Replicate Scatterplots

Scatterplots for E-scores, Z-scores, and the r2 values for each plot. 

1. Read the meta data and create output folder if needed.
2. Read in each pair of k-mer files and do an inner join on the k-mers. 
3. Draw a scatterplot for E-scores and save it.
4. Draw a scatterplot for Z-scores and save it.
5. Save the r2 values in the variable *regression_results*

**Input:** META_DATA_KMERS and the pairs of files reffered to by it.

**Output:** Scatterplots for E-scores and Z-scores in svg format.

#### (1) Meta data input:

```python
# Read and display kmer sequence meta data
kmer_meta_data = pd.read_csv(META_DATA_REP_KMERS)
kmer_meta_data
```

```python
# Read and display kmer plot parameter metadata
kmer_plot_meta_data = pd.read_csv(META_DATA_REP_KMERS_PLOT)
kmer_plot_meta_data
```

#### (2) Functions:

```python
def make_plot_ticks(start, end, step):
    """Return list of integers from a start, end, and step."""
    result = [start]
    while result[-1] < end:
        result.append(result[-1] + step)
    return result

def add_comparison_scatterplot(plot_df, params, ax):
    """Add a scatterplot to an ax object in a matplotlib figure."""
    sns.scatterplot(x=params.Comparison_X,
                    y=params.Comparison_Y,
                    data=plot_df,
                    ax=ax,
                    linewidth=0,
                    color="black",
                    legend=False,
                    s=12)
    ax.set_xlim(params.Range_Start, params.Range_End)
    ax.set_ylim(params.Range_Start, params.Range_End)
    plot_ticks = make_plot_ticks(params.Tick_Start, params.Tick_End, params.Tick_Step)
    ax_array[idx].set_xticks(plot_ticks)
    ax_array[idx].set_yticks(plot_ticks)
    ax_array[idx].set_xticklabels([])
    ax_array[idx].set_yticklabels([])
    ax_array[idx].set(adjustable='box', aspect='equal')
        
def calculate_rsqr(plot_df, params):
    """Return the r squared value and input information between 2 columns in a dataframe.
    
    Input is a dataframe with at least 2 columns of values and a namedtuple params which contains
    the following named parameters: Comparison_X, Comparison_Y, Protein, Metric. Returns a tuple
    with the protein, metric, x and y columns used in the calculation and the r squared value
    itself.
    """
    regression = scipy.stats.linregress(plot_df[params.Comparison_X], plot_df[params.Comparison_Y])
    r_sqr = regression.rvalue ** 2
    return (params.Protein, params.Metric, params.Comparison_X, params.Comparison_Y, r_sqr)
```

#### (3) Analysis:

```python
# Read k-mer dataframes and concatinate as replicate_kmers
kmer_dfs = []
for row in kmer_meta_data.itertuples():
    df = pd.read_csv(f"{KMER_FOLDER}/{row.File}", sep='\t')
    df["Protein"] = row.Protein
    df["Experiment"] = row.Experiment
    kmer_dfs.append(df)
replicate_kmers = pd.concat(kmer_dfs).reset_index(drop=True)

# Create a dictionary with pivoted dataframes for E-scores and Z-scores
score_df_dict = {}
for i in ("E-score", "Z-score"):
    score_df = replicate_kmers[["7-mer", i, "Protein", "Experiment"]]
    score_df = score_df.pivot(index=["7-mer", "Protein"],
                              columns="Experiment",
                              values=i)
    score_df = score_df.reset_index()
    score_df_dict[i] = score_df

# Make a scatterplot and save the s_sqr value across metric and protein
r_sqr_results = []
for groupby_cols, group_df in kmer_plot_meta_data.groupby(by=["Protein", "Metric"]):
    # Expand protein and metric from the groupby_cols
    protein, metric = groupby_cols
    # Subset the score dataframe by the protein being plotted and metric
    score_plot_df = score_df_dict[metric][score_df_dict[metric]["Protein"] == protein]
    # Set figure parameters based on the number of plots to draw
    comparison_count = len(group_df)
    fig, ax_array = plt.subplots(1, comparison_count)
    fig.set_size_inches(comparison_count ** 2, comparison_count)
    # For each comparison, add a scatterplot to the figure
    for idx, row in enumerate(group_df.itertuples()):
        add_comparison_scatterplot(score_plot_df, row, ax_array[idx])
        r_sqr_results.append(calculate_rsqr(score_plot_df, row))
    plt.subplots_adjust(wspace=0.35)
    # Save the figure in SVG format
    fig.savefig(fname=f"{OUTPUT_FOLDER}/{protein}_{metric}.svg", format='svg')

    
```

#### (1) Read and display meta data

#### (2) Functions and classes:

```python
def add_custom_comparison_scatterplot(plot_df, params, ax):
    """Add a scatterplot to an ax object in a matplotlib figure."""
    sns.scatterplot(x=params.Comparison_X,
                    y=params.Comparison_Y,
                    data=plot_df,
                    ax=ax,
                    linewidth=0,
                    color="black",
                    legend=False,
                    s=12)
    plot_range = list(map(np.log, [params.Range_Start, params.Range_End]))
    ax.set_xlim(plot_range[0], plot_range[1])
    ax.set_ylim(plot_range[0], plot_range[1])
    plot_ticks = list(map(np.log, [params.Tick_Low, params.Tick_Medium, params.Tick_High]))
    ax_array[idx].set_xticks(plot_ticks)
    ax_array[idx].set_yticks(plot_ticks)
    ax_array[idx].set_xticklabels([])
    ax_array[idx].set_yticklabels([])
    ax_array[idx].set(adjustable='box', aspect='equal')
    
    
PRIMER = uac.PRIMER



def return_group(x: str):
    """Return a substring representing the group a label belongs to."""
    # Determine how many subgroups deliminated by '_' to return
    n = 3
    if x.startswith("MITF_Ex"):
        n = 4
    elif x.startswith("random_damageable"):
        n = 2
    # Return the subgroup
    return '_'.join(x.split('_')[:n])


def filter_alldata_dataframe(dataframe: pd.DataFrame,
                             primer: str = PRIMER) -> pd.DataFrame:
    """Remove flagged sequence and non-double stranded sequences.

    Given a dataframe with column names in the alldata format, returns
    a dataframe with all flagged rows, rows with no sequence, and
    rows with a sequence without a primer removed.
    """
    alexa = dataframe.columns[7]
    # Filter out flags
    dataframe = dataframe[(dataframe["Cy3Flags"].isnull()) &
                          (dataframe[f"{alexa}Flags"].isnull())]
    # Remove rows with no sequence
    dataframe = dataframe[dataframe["Sequence"].notnull()]
    # Remove rows without primer
    dataframe = dataframe[dataframe["Sequence"].str.endswith(primer)]
    dataframe = dataframe.reset_index(drop=True)
    return dataframe


def replicate_values(dataframe: pd.DataFrame,
                     column: str,
                     min_count: int = 2) -> pd.DataFrame:
    """Return a dataframe of values with a minimum number of replicates.

    Given a dataframe and column, finds the number of replicates of that
    column and returns a dataframe with a minimum amount of replicates.
    """
    # Calculate value counts and return as a dataframe
    count = pd.DataFrame(dataframe[column].value_counts()).reset_index()
    # Rename the columns
    count = count.rename(columns={column: "Count"})
    count = count.rename(columns={"index": column})
    # Query for items that have a minimum value count
    count = count.query("Count >= @min_count")
    return count


def process_alldata_file(file_path: str) -> pd.DataFrame:
    """Given an alldata file, return a dataframe of custom sequences.

    Takes an alldata file and does the following pipeline:
    1. Read as a dataframe
    2. Remove universal sequences
    3. Remove trailing replicate labels so replicate data can be queried
    4. Subset the column into the sequence name, sequence, and signal values

    Returns a dataframe with a Name, Sequence, and Signal columns.
    """
    # Read dataframe
    dataframe = pd.read_csv(file_path, sep='\t')
    alexa = dataframe.columns[7]  # Name is the specific alexa wavelength for the experiment
    dataframe = filter_alldata_dataframe(dataframe)
    # Remove universal UV-Bind Probes
    dataframe = dataframe[~dataframe["Name"].str.startswith("All")]
    dataframe = dataframe.reset_index(drop=True)
    # Remove trailing replicate label for replicate counting
    dataframe["Name_Group"] = dataframe["Name"]\
        .apply(lambda x: '_'.join(x.split('_')[:-1]))
    # Filter for sequences with 2+ replicates
    replicate_counts = replicate_values(dataframe, "Name_Group")
    name_set = set(replicate_counts["Name_Group"])
    dataframe = dataframe[dataframe["Name_Group"]
                          .isin(name_set)].reset_index(drop=True)
    # Subset by name, sequence and signal columns
    dataframe = dataframe[["Name", "Sequence", alexa]]
    dataframe = dataframe.rename(columns={alexa: "Signal"})
    return dataframe

def filter_rep(x):
    """Remove the _r suffix from a label."""
    if re.search(pattern="_r\d*$", string=x):
        return '_'.join(x.split('_')[:-1])
    else:
        return x
    
def decompress_tbp_labels(df):
    dup = df[df["Name"].apply(lambda x: "Ariel_A" in x and "Ariel_G" in x)].copy()
    dup = dup.reset_index(drop=True)
    dup["Group"] = "TBP_Ariel_G"
    df = pd.concat([df, dup]).reset_index(drop=True)
    return df
```

#### (3) Prepare data for plotting

```python
# Read dataframes indicated in the meta data
dfs = []
for row in custom_meta.itertuples():
    df = process_alldata_file(f"{ALLDATA_FOLDER}/{row.File}")
    df["Group"] = df["Name"].apply(lambda x: x.split('-')[0])
    df["Group"] = df["Group"].apply(lambda x: filter_rep(x))
    df = decompress_tbp_labels(df)
    df["Name"] = df["Name"].apply(lambda x: "_".join(x.split("_")[:-1]))
    df["Sequence"] = df["Sequence"].apply(lambda x: x[:25])
    df = df.groupby(by=["Group","Name", "Sequence"]).aggregate(np.median).reset_index()
    df["Experiment"] = row.Label
    dfs.append(df)
replicate_custom_df = pd.concat(dfs).reset_index(drop=True)
replicate_custom_df = replicate_custom_df[replicate_custom_df["Group"].isin(PDB_VAR_LIB)]
# Convert to ln space
replicate_custom_df["Signal"] = replicate_custom_df["Signal"].apply(lambda x: np.log(x))
replicate_custom_df = replicate_custom_df.pivot(index=["Group", "Name", "Sequence"],
                                                values = "Signal",
                                                columns="Experiment").reset_index()
replicate_custom_df
```

#### (4) Create Scatterplots

```python
for protein, group_df in custom_plotting_meta.groupby(by="Protein"):
    # Subset the score dataframe by the protein being plotted and metric
    plot_df = replicate_custom_df
    # Set figure parameters based on the number of plots to draw
    comparison_count = len(group_df)
    fig, ax_array = plt.subplots(1, comparison_count)
    fig.set_size_inches(comparison_count ** 2, comparison_count)
    # For each comparison, add a scatterplot to the figure
    for idx, row in enumerate(group_df.itertuples()):
        add_custom_comparison_scatterplot(plot_df, row, ax_array[idx])
        r_sqr_results.append(calculate_rsqr(plot_df, row))
    plt.subplots_adjust(wspace=0.35)
    # Save the figure in SVG format
    fig.savefig(fname=f"{OUTPUT_FOLDER}/{protein}_custom.svg", format='svg')
    
```

#### (5) Save R2 Results

Store the results from regression_df as a dataframe and save to a table.

```python
regression_df = pd.DataFrame(r_sqr_results)
regression_df = regression_df.rename(columns={0:"Protein", 1:"Metric", 2:"Experiment_X",3:"Experiment_Y", 4:"R_Sqr"})
regression_df.to_csv(f"{OUTPUT_FOLDER}/R_Sqr_Values.csv", float_format='%g')
regression_df
```
