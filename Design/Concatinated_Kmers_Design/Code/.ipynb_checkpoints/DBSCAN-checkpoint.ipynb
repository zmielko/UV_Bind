```python
import pandas as pd
import numpy as np
from sklearn.cluster import dbscan
from jellyfish import levenshtein_distance
import argparse

def argumentParse():
    programDescription = 'Affinity Propagation of Probes'
    parser = argparse.ArgumentParser(description=programDescription)
    req = parser.add_argument_group('required arguments')
    opt = parser.add_argument_group('optional arguments')
    req.add_argument('-i','--input',required = True, type=str, help='Input file')
    req.add_argument('-oC','--outputCluster',required = True, type=str, help='Output cluster file as tsv')
    req.add_argument('-oM','--outputMatrix', required = True, type=str,help='Output matrix file as npz')
    args = parser.parse_args()
    return(args)



def levensteinMatrix(stringArray):
    """
    Given an array of strings, returns the levenshtein distance for 
      each string to every other string as a matrix. 
    """
    listDistList = []
    for string1 in stringArray:
        distList = []
        for string2 in stringArray:
            distList.append(levenshtein_distance(string1, string2))
        listDistList.append(distList)
    distMatrix = np.array(listDistList)
    return(distMatrix)

if __name__ == '__main__':
    # Parse arguments
    args = argumentParse()
    # Read input file, convert to array
    print(f"ARGS_INPUT = {args.input}")
    df = pd.read_csv(args.input, sep = '\t', header = None)
    seqList = list(df[0])
    # Construct matrix of levenstein distances, save
    lMatrix = levensteinMatrix(df[0])
    np.savez_compressed(args.outputMatrix, lMatrix, delimiter='\t')
    # Convert all distances to negative values
    #  Larger distances incur a greater penalty
    lMatrix = lMatrix * -1
    #X = np.arange(len(seqList)).reshape(-1, 1)
    cluster = dbscan(lMatrix, metric="precomputed", eps=5, min_samples=3)
    # Add cluster labels to input DataFrame and save
    df["Cluster"] = cluster[1]
    df = df.rename(columns={0:"Sequence"})
    df = df.sort_values(by = "Cluster")
    df.to_csv(args.outputCluster, sep = '\t', index = False)
```

```python
test = "/Users/ZMielko/Desktop/UV_TBP_MITF_EGR1_CREB1_Array/Data/Diff_UV_Probes/EGR1_extShort.tsv"
testData = pd.read_csv(test, sep = '\t', header = None)
testData = testData[0]
def levensteinMatrix(stringArray):
    """
    Given an array of strings, returns the levenshtein distance for 
      each string to every other string as a matrix. 
    """
    listDistList = []
    for string1 in stringArray:
        distList = []
        for string2 in stringArray:
            LD = levenshtein_distance(string1, string2)
            if LD <= 3:
                distList.append(1)
            else:
                distList.append(0)
        listDistList.append(distList)
    distMatrix = np.array(listDistList)
    return(distMatrix)
lMatrix = levensteinMatrix(testData)
lMatrix
```

```python
import markov_clustering as mc
import networkx as nx
import random

# number of nodes to use
numnodes = 200

# generate random positions as a dictionary where the key is the node id and the value
# is a tuple containing 2D coordinates
positions = {i:(random.random() * 2 - 1, random.random() * 2 - 1) for i in range(numnodes)}

# use networkx to generate the graph
network = nx.random_geometric_graph(numnodes, 0.3, pos=positions)

# then get the adjacency matrix (in sparse form)
matrix = nx.to_scipy_sparse_matrix(network)
```
