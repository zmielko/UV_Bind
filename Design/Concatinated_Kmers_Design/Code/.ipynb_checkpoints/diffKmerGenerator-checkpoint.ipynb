Algorithm that extends a site to meet the 25bp needed for a probe.

Filters by WC E-score < 0 and then does a greedy extension for highest UV E-score. First extends to the left one base, then the right and keeps going until 25bp are met or it cannot continue. 

```python
kDictNonSpecDiff
```

```python
# Algorithm that extends differential k-mer sequences
nonSpecWCSet = set(list(nonSpecWC["kmerFwd"]) + list(nonSpecWC["kmerRC"]))

def extCaniR(seq, k=6, kSet = nonSpecWCSet):
    queryStart = len(seq) - k + 1
    querySeq = seq[queryStart:]
    bestCani = ''
    bestCaniPref = -0.5
    for i in ['A', 'C', 'G', 'T']:
        if querySeq + i in kSet:
            if kDictNonSpecDiff[querySeq + i] > bestCaniPref:
                bestCani = i
                bestCaniPref = kDictNonSpecDiff[querySeq + i]
    return(seq + bestCani)
curSeq = "GGAATCCCG"
for i in range(5):
    curSeq = extCaniR(curSeq)
curSeq
```

```python
def extCaniL(seq, k=6, kSet = nonSpecWCSet):
    querySeq = seq[:k - 1]
    bestCani = ''
    bestCaniPref = -0.5
    for i in ['A', 'C', 'G', 'T']:
        if i + querySeq in kSet:
            if kDictNonSpecDiff[i + querySeq] > bestCaniPref:
                bestCani = i
                bestCaniPref = kDictNonSpecDiff[i + querySeq]
    return(bestCani + seq)

curSeq = "GGAATCCCG"
for i in range(5):
    curSeq = extCaniL(curSeq)
curSeq
```

```python
def extend25(seq):
    extProbe = seq
    prevLen = len(extProbe)
    while True:
        extProbe = extCaniL(extProbe)
        if len(extProbe) == 25:
            return(extProbe)
        extProbe = extCaniR(extProbe)
        if len(extProbe) == 25:
            return(extProbe)
        if len(extProbe) == prevLen:
            print("Cannot extend probe!")
            print(extProbe)
            print(len(extProbe))
            return("N")
        prevLen = len(extProbe)
extend25("CGATAAA")
```
