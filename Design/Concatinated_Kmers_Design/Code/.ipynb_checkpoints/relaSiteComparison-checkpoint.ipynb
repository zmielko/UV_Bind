```python
import pandas as pd
import numpy as np
fileTup = [(f"EGR1_WLS/Escore_6_8_vs_Escore_6_9_UV_CI_0.95.txt", "EGR1_WLS/EGR_kmers.txt", "EGR1"),
           (f"CREB1_WLS/Escore_6_8_vs_Escore_6_9_UV_CI_0.95.txt", "CREB1_WLS/CREB_kmers.txt", "CREB1"),
           (f"TBP_WLS/Escore_6_8_vs_Escore_6_9_UV_CI_0.95.txt", "TBP_WLS/TBP_kmers.txt", "TBP"),]
dataFolder = "../Data"
fT = 1
wls = pd.read_csv(f"{dataFolder}/{fileTup[fT][0]}", sep = '\t')
kwls = pd.read_csv(f"{dataFolder}/{fileTup[fT][1]}", sep = '\t')
wls["kmer"] = kwls["kmerFwd"]
wls["kmerR"] = kwls["kmerRC"]
wls["Diff"] = wls["Escore_6_9_UV"] - wls["Escore_6_8"]
kDictScoreDiff = dict(zip(wls["kmer"], wls["Diff"])) 
kDictScoreDiff.update(dict(zip(wls["kmerR"], wls["Diff"])))
nonSpecWC = kwls.query("Escore_6_8 < 0").copy(deep = True)
nonSpecWC["Diff"] = nonSpecWC["Escore_6_9_UV"] - nonSpecWC["Escore_6_8"]
nonSpecWCSet = set(list(nonSpecWC["kmerFwd"]) + list(nonSpecWC["kmerRC"]))
kDictNonSpecDiff = dict(zip(nonSpecWC["kmerFwd"], nonSpecWC["Diff"])) 
kDictNonSpecDiff.update(dict(zip(nonSpecWC["kmerRC"], nonSpecWC["Diff"])))
UVpref = wls.query("pref < 0")
prefSet = set(list(UVpref["kmer"]) + list(UVpref["kmerR"]))
```

```python
def sumScore(seq, k, kScoreDict):
    score = 0
    for i in range(len(seq) - k + 1):
        score = score + kScoreDict[seq[i:i+k]]
    return(score)
```

```python
gen25 = pd.read_csv("../Data/diffUVProbes/EGR1_25mers.tsv", sep = '\t', header = None)
gen25["scores"] = gen25[0].apply(lambda x: sumScore(x, 6, kDictScoreDiff))
gen25.sort_values(by = "scores", ascending = False)
```

```python
rscore = sumScore("GCGGAAATTCCCGGGAATTTCCGCG", 6, kDictScoreDiff)
gen25.query("scores > @rscore")
```

```python
from jellyfish import hamming_distance
def closestSubHamming(string, query):
    qLen = len(query)
    minHam = qLen
    for i in range(len(string) - qLen + 1):
        ham = hamming_distance(query, string[i:i+qLen])
        if ham < minHam:
            minHam = ham
    return(minHam)
hDist = []
for i in gen25[0]:
    hDist.append(closestSubHamming(i, "GGAAATTCCC"))
gen25["RA_hdist"] = hDist
gen25.sort_values(by = "RA_hdist")
```

```python
from jellyfish import levenshtein_distance
RELA = "GCGGAAATTCCCGGGAATTTCCGCG"
lDist = []
for i in gen25[0]:
    lDist.append(levenshtein_distance(RELA, i))
gen25["RA_ldist"] = lDist
gen25.sort_values(by = "RA_ldist")
```
